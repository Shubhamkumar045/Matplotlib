```python
Matplotlib

Q1: What is Matplotlib? Why is it used? Name five plots
that can be plotted using the Pyplot module of
Matplotlib.
Answer--Matplotlib is a comprehensive library for creating static, interactive, 
and animated visualizations in Python. It provides a wide range of functionalities
to generate high-quality plots, charts, and figures for various purposes.
Matplotlib is extensively used in fields such as data science, machine learning,
scientific computing, and more.
Five common plots that can be created using the Pyplot module of Matplotlib are:

Line Plot: Displays data points connected by straight lines.
It is commonly used to visualize trends and patterns over continuous data.

Scatter Plot: Represents individual data points as markers on a 2D plane.
It is useful for visualizing the relationship between two variables
and identifying patterns, clusters, or outliers.

Histogram: Shows the distribution of a continuous variable by dividing the
data into bins and displaying the frequency of observations in each bin. 
Histograms are useful for understanding the underlying distribution of data.

Bar Plot: Represents categorical data using rectangular bars with heights or 
lengths proportional to the values they represent. Bar plots are suitable for
comparing values across different categories.

Pie Chart: Displays data as a circular chart divided into sectors, where each
sector represents a proportion of the whole. Pie charts are useful for visualizing 
the relative contributions of different categories to a total.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
Answer--import numpy as np
import matplotlib.pyplot as plt

# Generating data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Creating a scatter plot
plt.scatter(x, y, color='blue', alpha=0.7)  # alpha sets transparency for better visibility

# Adding title and labels
plt.title('Scatter Plot of x and y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Displaying the plot
plt.grid(True)  # Adding grid for better readability
plt.show()


Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
Answer--The subplot() function in Matplotlib is used to create multiple plots within the same figure.
It allows you to arrange plots in a grid-like fashion, specifying the number of rows and columns
and the position of each subplot within the grid.

import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots and drawing line plots
plt.figure(figsize=(12, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, color='blue')
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, color='green')
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, color='red')
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, color='purple')
plt.title('Line Plot 4')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()


Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
Answer--A bar plot is a type of graph that presents categorical data with rectangular bars. 
The lengths of the bars are proportional to the values they represent.
Bar plots are commonly used to compare the values of different categories
or to show the distribution of data across categories.

why bar plots are used:

A)Comparison: Bar plots provide a visually intuitive way to compare the values of different categories or groups.
B)Distribution: They can show the distribution of data across categories, especially when dealing with discrete variables.
C)Easy Interpretation: Bar plots are easy to interpret and understand, making them suitable for presentations and reports.


import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')

# Adding title and labels
plt.title('Profit of Companies')
plt.xlabel('Company')
plt.ylabel('Profit (in $)')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)  # Add grid lines for better readability
plt.tight_layout()
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')

# Adding title and labels
plt.title('Profit of Companies')
plt.xlabel('Profit (in $)')
plt.ylabel('Company')

# Show the plot
plt.grid(True)  # Add grid lines for better readability
plt.tight_layout()
plt.show()


Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Answer--A box plot, also known as a box-and-whisker plot, is a statistical
visualization method used to display the distribution and variability of a dataset. 
It provides a concise summary of the central tendency, dispersion, and 
skewness of the data, as well as the presence of outliers.

why box plots are used:

Summary of Distribution: Box plots provide a visual summary of the distribution of the data,
including the median, quartiles, and range.
Identification of Outliers: They help in identifying outliers, which are data points that lie 
significantly outside the overall pattern of the data.
Comparison: Box plots allow for easy comparison of the distributions of multiple datasets or groups.\
```
